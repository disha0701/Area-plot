```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame
```

```python
df = pd.read_excel('./downloads/USA.xlsx',
                  sheet_name = 'USA by Place of birth',
                  skiprows=range(20),
                   skipfooter=2)
df.head()
```

```python
df.columns.values
```

```python
df.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df.head(7)
```

```python
#To check to see how many null objects we have in the dataset as follows

df.isnull().sum()
```

```python
df.set_index('OdName', inplace=True)
df
```

```python

df['Total'] = df.sum(axis=1)
df
```

```python
df.columns = list(map(str, df.columns))
years = list(map(str, range(1980, 2014)))
years
df_CI = df.loc[['India','China'], years] # passing in years 1980 - 2013 to exclude the 'total' column
df_CI.head()
df_CI.index.name = None
df
```

```python

df_CI = df_CI.transpose()
df_CI.head()
```

```python
df_CI.index = df_CI.index.map(int)  
df_CI.plot()
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
```

```python
df.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df
```

```python

df.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df.head(5)
df_top5 = df_top5[years].transpose() 
print(df_top5)

df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top5.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size

plt.show()
```
